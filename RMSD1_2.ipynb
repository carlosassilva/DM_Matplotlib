```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = [], []

with open("rmsd.xvg") as df:
    for line in df:
        cols = line.split()

        if len(cols) == 2:
            x.append(float(cols[0]))
            y.append(float(cols[1]))
```

```python
a, b = [], []

with open("rmsd2.xvg") as df:
    for line in df:
        cols = line.split()

        if len(cols) == 2:
            a.append(float(cols[0]))
            b.append(float(cols[1]))
```

```python
#RMSD_Plot_1
fig = plt.figure()
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)
#Data_RMSD1
ax.fill_between(x, y, color="deepskyblue", linestyle="-", alpha=0.1)
ax.plot(x, y, color="blue", linestyle="-", label='Vv-Def1', alpha=0.9)
#Data_RMSD2
ax.fill_between(a, b, color="deepskyblue", linestyle="-", alpha=0.1)
ax.plot(a, b, color="red", linestyle="-", label='Vv-Def2', alpha=0.9)
#Legend
ax.set_xlabel("Time(ns)")
ax.set_ylabel("RMSD (nm)")
ax.set_title("RMSD") 
leg = ax.legend()
#Axes
plt.xlim(0, 100)
plt.ylim(0, 0.25)
ax.tick_params(axis ='both', which ='both', length = 5)
#Save_plot
plt.savefig('RMSD1_2.png', dpi=300)
plt.savefig("RMSD1_2.svg")
plt.savefig("RMSD1_2.pdf")
#Display
plt.show() 
```
