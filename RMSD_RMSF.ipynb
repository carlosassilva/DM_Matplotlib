```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
x, y = [], []

with open("rmsd.xvg") as df:
    for line in df:
        cols = line.split()

        if len(cols) == 2:
            x.append(float(cols[0]))
            y.append(float(cols[1]))
```

```python
#RMSD_Plot_1
fig = plt.figure()
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)
ax.fill_between(x, y, color="deepskyblue", linestyle="-", alpha=0.2)
ax.plot(x, y, color="blue", linestyle="-", label='RMSD', alpha=0.9)
#Legend
ax.set_xlabel("Time(ps)")
ax.set_ylabel("RMSD (nm)")
ax.set_title("RMSD") 
leg = ax.legend()
#Axes
plt.xlim(0, 100000)
plt.ylim(0, 0.7)
ax.tick_params(axis ='both', which ='both', length = 5)
#Save_plot
plt.savefig('RMSD_1.png', dpi=300)
plt.savefig("RMSD_1.svg")
plt.savefig("RMSD_1.pdf")
#Display
plt.show() 
```

```python
#RMSD_Plot_2
fig, (ax) = plt.subplots(1, 1, sharex=True)

ax.fill_between(x, y, color="deepskyblue", linestyle="-", alpha=0.2)
ax.plot(x, y, color='black', linestyle=":", alpha=0.9)
#Legend
ax.set_xlabel("Time(ps)")
ax.set_ylabel(r"RMSD (nm)")
ax.set_title("RMSD") 

#Axes
plt.xlim(0, 100000)
plt.ylim(0, 0.75)
ax.tick_params(axis ='both', which ='both', length = 5)
#plt.xticks(np.arange(min(x), max(x), 10000))
plt.xticks(np.arange(0, 110000, (10000)))
plt.xticks(rotation=45)
#Save_plot
plt.savefig('RMSD_2.png', dpi=300)
plt.savefig("RMSD_2.svg")
plt.savefig("RMSD_2.pdf")
#Display
plt.show() 
```

```python
df_xy = pd.read_csv("rmsf_2.csv", sep =";")
a = df_xy ["x"]
b = df_xy ["y"]
```

```python
#RMSF_lollipop
markers, stemlines, baseline = plt.stem(a, b, use_line_collection=True, linefmt='darkorchid')
plt.setp(markers, marker='D', markersize=10, markeredgecolor="black", markeredgewidth=2)
#Legend
plt.xlabel("Residue")
plt.ylabel(r"C$_\alpha$ RMSF")
plt.title("RMS Fluctuation")
#Axes
plt.xlim(0, 21)
plt.ylim(0, 0.7)
plt.xticks(a)
#Save_plot
plt.savefig('RMSF_1.png', dpi=300)
plt.savefig("RMSF_1.svg")
plt.savefig("RMSF_1.pdf")
#Display
plt.show()
```
