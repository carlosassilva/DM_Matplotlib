```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = [], []

with open("rmsf.xvg") as df:
    for line in df:
        cols = line.split()

        if len(cols) == 2:
            x.append(float(cols[0]))
            y.append(float(cols[1]))
```

```python
a, b = [], []

with open("rmsf2.xvg") as df:
    for line in df:
        cols = line.split()

        if len(cols) == 2:
            a.append(float(cols[0]))
            b.append(float(cols[1]))
```

```python
#RMSF_lollipop
markers, stemlines, baseline = plt.stem(a, b, use_line_collection=True, linefmt='darkorchid')
plt.setp(markers, marker='o', markersize=10, markeredgecolor="black", markeredgewidth=1.2)
#Legend
plt.xlabel("Residue")
plt.ylabel(r"C$_\alpha$ RMSF")
plt.title("RMS Fluctuation")
#Axes
plt.ylim(0, 0.3)
plt.xticks(np.arange(1, 50, 5))
#Save_plot
plt.savefig('RMSF1.png', dpi=300)
plt.savefig("RMSF1.svg")
plt.savefig("RMSF1.pdf")
#Display
plt.show()
```

```python
#RMSF_lollipop
markers, stemlines, baseline = plt.stem(x, y, use_line_collection=True, linefmt='darkcyan')
plt.setp(markers, marker='D', markersize=8, markeredgecolor="black", markerfacecolor="red", markeredgewidth=1.2)
#Legend
plt.xlabel("Residue")
plt.ylabel(r"C$_\alpha$ RMSF")
plt.title("RMS Fluctuation")
#Axes
plt.ylim(0, 0.30)
plt.xticks(np.arange(1, 50, 5))
#Save_plot
plt.savefig('RMSF2.png', dpi=300)
plt.savefig("RMSF2.svg")
plt.savefig("RMSF2.pdf")
#Display
plt.show()
```

```python
#RMSF_1
markers, stemlines, baseline = plt.stem(a, b, use_line_collection=True, linefmt='cornflowerblue')
plt.setp(markers, marker='D', markersize=7, markeredgecolor="black", markeredgewidth=1.2)
#RMSF_2
markers, stemlines, baseline = plt.stem(x, y, use_line_collection=True, linefmt='cornflowerblue')
plt.setp(markers, marker='^', markersize=9, markeredgecolor="black",markerfacecolor="red", markeredgewidth=1.2)
#Legend
plt.xlabel("Residue")
plt.ylabel(r"C$_\alpha$ RMSF")
plt.title("RMS Fluctuation")
#Axes
plt.ylim(0, 0.3)
plt.xticks(np.arange(1, 50, 5))
#Save_plot
plt.savefig('RMSF1_2.png', dpi=300)
plt.savefig("RMSF1_2.svg")
plt.savefig("RMSF1_2.pdf")
#Display
plt.show()
```
